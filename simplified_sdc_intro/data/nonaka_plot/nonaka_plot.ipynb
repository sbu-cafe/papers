```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sdc = np.genfromtxt("nonaka_sdc_1210.dat", names=True)
strang = np.genfromtxt("nonaka_strang_1210.dat", names=True)
```

```python
sdc["sdciter"]
```

```python
fig, ax = plt.subplots()

idx0 = sdc["sdciter"] == 0
idx1 = sdc["sdciter"] == 1

ax.scatter(strang["time"], strang["Xni56"], marker="o", label="Strang")
ax.scatter(sdc["time"][idx0], sdc["Xni56"][idx0], marker="^", label="SDC iter 1")
ax.scatter(sdc["time"][idx1], sdc["Xni56"][idx1], marker="x", label="SDC iter 2")

ax.set_xlabel("time")
ax.set_ylabel("X(Ni56)")
ax.legend()
fig.set_size_inches((8, 8))
fig.savefig("sdc_compare.png", dpi=100)
```
