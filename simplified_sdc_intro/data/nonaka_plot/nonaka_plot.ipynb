```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sdc = np.genfromtxt("nonaka_sdc_1210.dat", names=True)
strang = np.genfromtxt("nonaka_strang_1210.dat", names=True)
```

```python
sdc["sdciter"]
```

```python
strang["stranghalf"]
```

```python
fig, ax = plt.subplots()
```

```python
def plot(ax, strang, sdc, var="Xni56", label=r"X(Ni56)"):
    idx0 = strang["stranghalf"] == 0
    idx1 = strang["stranghalf"] == 1

    ax.scatter(strang["time"][idx0], strang[var][idx0], marker="o", label="Strang first half")
    ax.scatter(strang["time"][idx1], strang[var][idx1], marker="*", label="Strang second half")

    idx0 = sdc["sdciter"] == 0
    idx1 = sdc["sdciter"] == 1

    ax.scatter(sdc["time"][idx0], sdc[var][idx0], marker="^", label="SDC iter 1")
    ax.scatter(sdc["time"][idx1], sdc[var][idx1], marker="x", label="SDC iter 2")

    ax.set_xlabel("time")
    ax.set_ylabel(label)
    ax.legend()
  
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1)
```

```python
plot(ax[0], strang, sdc, "Xni56", "X(Ni56)")
plot(ax[1], strang, sdc, "temperature", "T")
plot(ax[2], strang, sdc, "density", r"$\rho$")
fig
```
